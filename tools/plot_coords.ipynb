# Script to return PC coordinates for LaTeX-plot generation 

```python
import numpy as np
import os
```

```python
os.getcwd()
path = os.getcwd().split('/')
path_lvl_up = '/'.join(path[:-1])
data_path = path_lvl_up + '/data'
```

```python
data = np.loadtxt(data_path + '/pca_train.csv',delimiter=',')
```

```python
labels = data[:,-1]
```

## Choose two components to plot against

```python
components = [4,5]
```

```python
xy = np.column_stack((data[:,components[0]-1],data[:,components[1]-1]))
```

```python
plot_data = np.column_stack((xy,labels))
```

```python
plot_data
```

```python
classes = {1:'a',2:'b',3:'c',5:'d',6:'e',7:'f'}


for coord in plot_data:
    print(round(coord[0],3), round(coord[1],3), classes[coord[2]])
```
